Day 19: How do I select multiple rows and columns from a pandas DataFrame?


```python
import pandas as pd
```

```python
ufo = pd.read_csv('https://bit.ly/uforeports')
```

```python
ufo.head()
```

loc is a dataframe method - selecting rows by index and columns by label. 

```python
ufo.loc[0, :] # This is 0 row and all columns, : means all columns. This will return a dataframe series.
```

```python
ufo.loc[[0,1,2], :] # Here we want rows 1,2,3 but with all columns.
```

```python
ufo.loc[0:2, :] # Here we want rows 1,through to 3 inclusive on both side, but with all columns.
```

```python
ufo.loc[0:2] # This is same as ufo.loc[0:2, :] as Pandas will assume it.
```

```python
ufo.loc[:,'City'] # All row and city column.
```

```python
ufo.loc[:,['City', 'State']] # All row and city and state column.
```

```python
ufo.loc[:,'City': 'State'] # Range of columns
```

```python
ufo.loc[0:2,'City': 'State']
```

```python
ufo.head(3).drop('Time', axis=1)
```

```python
ufo[ufo['City'] == 'Oakland'] # We can also do ufo[ufo.City=='Oakland'].State
# If you want to select all rows where city name is oakland. 
```

```python
ufo.loc[ufo['City'] == 'Oakland', :] # Here we want only the rows that has city name as oakland and all columns.
```

```python
ufo.loc[ufo['City'] == 'Oakland', 'State'] 
```

```python
# iloc selecting rows and columns using integer positions [loc is with labels]
ufo.iloc[:,[0,3]] # position at integer value 0 and 3
```

```python
ufo.iloc[:, 0:4] # using iloc this is exclusive of the second limit, and inclusive of the first limit. Whereas in case of loc it is inclusive of the limit. 
```

```python
ufo.iloc[0:3, :] # This is rows 0,1,2 excluding row 3, and all columns.
```

```python
ufo[['City', 'State']] # The inner bracket is the list
```

```python
# We can accomplish the same using loc. 
ufo.loc[:, ['City', 'State']]
```

```python
ufo[0:2] # This is representing row with index.
```

```python
ufo.iloc[0:2, :]
```

```python
# ix - It allows you mix labels and integer i.e. it is a blend with loc and iloc
drinks = pd.read_csv('https://bit.ly/drinksbycountry', index_col='country')
drinks.head()
```

```python
# For selecting a cell - using label and position. 
drinks.ix['Afghanistan', 0]
```
